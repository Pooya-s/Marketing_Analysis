````python
import pandas as pd

file_path = 'Octopus_cleaned_data.csv'
data = pd.read_csv(file_path)
````

## 1. Analyze Positive Reviews
We'll extract insights from the ```Sentiment``` column and ```Cleaned_Text``` for highly rated reviews (4- and 5-star ratings) to identify recurring positive themes.

### Step 1: Filter Positive Reviews
We'll filter the dataset for reviews with ratings of 4 or 5.

### Step 2: Analyze Sentiment Scores
Among these high-rated reviews, examine the ```Sentiment``` column to ensure we focus on truly positive reviews.

### Step 3: Extract Text Data
Use the ```Cleaned_Text``` column to analyze themes. We'll tokenize the text and calculate the frequency of words or phrases.

### Step 4: Identify Themes
Categorize frequently mentioned words into themes such as:

* Customer Service: Look for words like "support," "help," "team."
* Pricing: Identify mentions of "affordable," "price," "value."
* Features: Highlight specific strengths like "easy," "efficient."

### Step 5: Present Insights
Summarize the recurring themes into actionable insights to craft promotional content.

````python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Step 1
positive_reviews = data[(data['Rating'] >= 4) & (data['Sentiment'] >= 0.27)]

# Step 2

# Step 3
all_positive_text = " ".join(positive_reviews['Cleaned_Text'].dropna())

words = all_positive_text.split()
word_counts = Counter(words)

# Step 4
top_words = word_counts.most_common(20)
words = [word for word, _ in top_words]
frequencies = [freq for _, freq in top_words]

plt.figure(figsize=(12, 6))
colors = ['#F09EA7' if i % 2 == 0 else '#C1E7E3' for i in range(len(words))]
bars = plt.bar(words, frequencies, color=colors)
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 10,
        int(bar.get_height()),
        ha='center', va='bottom', fontsize=10, color='black'
    )
plt.title("Top 20 Words in Positive Reviews", fontsize=16)
plt.xlabel("Words", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45, fontsize=10, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

### **Insights from Positive Reviews:**
The top 20 most frequent words reveal several key strengths highlighted in positive reviews:

* **Customer Service Excellence:**
 * Words like "customer," "service," "kind," and "thanks" suggest that customers highly value the customer support they receive.
Specific mentions of "operator" and "whatsapp" indicate that the ease of interaction is a notable strength.
* **Efficiency:**
 * Words like "fast"  and "always" highlight the efficiency of the service, which resonates with customers.
* **Clarity and Transparency:**
 * Frequent mentions of "clear," "bill," and "rate" suggest that customers appreciate the clarity and transparency in billing and pricing.

**Positive Brand Perception:**

The frequent use of terms like "excellent," "great," and "thank" showcases high satisfaction and positive sentiment toward the brand.

**Recommendations for Marketing Messages:**

1. Highlight Service Excellence: for example it would be siad: "Our award-winning customer service team is here for you—fast, friendly, and reliable."
2. Emphasize Efficiency: for example it would be siad: "Switch to Octopus Energy and experience quick, hassle-free support every time."
3. Reinforce Transparency:In this case, it would be emphesized in this way:  "No surprises, no hidden fees—clear and transparent billing you can trust."

## 2. Examine Negative Reviews
Focus on reviews with low ratings (e.g., 1- and 2-star) or low sentiment scores.

Identify recurring issues or areas of dissatisfaction, such as:
* Service delays or poor response times.
* Billing problems or perceived pricing issues.
* Regional or demographic-specific complaints.


````python
negative_reviews = data[(data['Rating'] <= 2) | (data['Sentiment'] < 0)]

all_negative_text = " ".join(negative_reviews['Cleaned_Text'].dropna())
words = all_negative_text.split()
word_counts = Counter(words)


top_words = word_counts.most_common(25)
words = [word for word, _ in top_words]
frequencies = [freq for _, freq in top_words]

plt.figure(figsize=(12, 6))
colors = ['#BDB2FF' if i % 2 == 0 else '#FEC868' for i in range(len(words))]
bars = plt.bar(words, frequencies, color=colors)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 2,
        int(bar.get_height()),
        ha='center', va='bottom', fontsize=10, color='black'
    )

plt.title("Top 20 Words in Negative Reviews", fontsize=16)
plt.xlabel("Words", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45, fontsize=10, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

### **Most Common Complaints**
- **Customer Service Issues**:
  - Words like "customer," "service," and "operator" suggest dissatisfaction with customer interactions.
  - These could indicate poor response times, lack of support, or unresolved issues.

- **Billing and Pricing Problems**:
  - Frequent mentions of "bill," "cost," "price," and "contract" point to issues related to billing transparency or unexpected charges.
  - Customers may feel overcharged or confused about their bills.

- **Company Reputation**:
  - Words like "company" and "octopus" being frequent could indicate a focus on the company itself, possibly tied to negative brand perception or poor experiences.

### **Regional or Supplier-Specific Complaints**
- Words like "supplier" and "energy" suggest potential issues with availability, quality, or reliability in energy supply.
- There might be regional variations in service quality.

### **Uncertainty or Complexity**
- Words like "email," "problem," "request," and "serious" suggest communication issues or unresolved customer queries.
- These could indicate a need for improved clarity and responsiveness in handling customer requests.

### **Actionable Recommendations**
1. **Customer Service Improvements**:

2. **Billing Transparency**:

## 4. Segment Marketing Campaigns
* **By Geography:**
 * If specific locations show unique strengths or weaknesses, craft tailored messaging for those regions.
* **By Customer Needs:**
 * Highlight features relevant to specific demographics.


````python
#Segment by Geography
geo_analysis = data.groupby('Location').agg(
    Average_Rating=('Rating', 'mean'),
    Average_Sentiment=('Sentiment', 'mean'),
    Total_Reviews=('Rating', 'count')
).reset_index()
geo_analysis = geo_analysis[geo_analysis['Location'] != 'Unknown']

high_performing_regions = geo_analysis[geo_analysis['Average_Rating'] >= 4]
low_performing_regions = geo_analysis[geo_analysis['Average_Rating'] < 4.0]

#Segment by Customer Needs
needs_keywords = {
    "Affordability": ["affordable", "price", "cost", "value", "save", "bill", "high", "low"],
    "Efficiency": ["efficient", "quick", "fast", "easy", "reliable", "slow", "terrible"],
    "Support": ["help", "support", "team", "operator", "service", "customer"]
}

# Count occurrences of keywords in reviews for segmentation
needs_counts = {need: 0 for need in needs_keywords}

for text in data['Cleaned_Text'].dropna():
    for need, keywords in needs_keywords.items():
        if any(keyword in text for keyword in keywords):
            needs_counts[need] += 1

geo_analysis_high = high_performing_regions.sort_values(by='Average_Rating', ascending=False)
geo_analysis_low = low_performing_regions.sort_values(by='Average_Rating')

needs_analysis = pd.DataFrame.from_dict(needs_counts, orient='index', columns=['Count']).reset_index()
needs_analysis.rename(columns={'index': 'Customer_Need'}, inplace=True)
````

````python
print("High-Performing Regions:\n")
if not geo_analysis_high.empty:
    print(geo_analysis_high.to_string(index=False))
else:
    print("No high-performing regions found.\n")

print("\nLow-Performing Regions:\n")
if not geo_analysis_low.empty:
    print(geo_analysis_low.to_string(index=False))
else:
    print("No low-performing regions found.\n")

print("\nCustomer Needs Analysis:\n")
if not needs_analysis.empty:
    print(needs_analysis.to_string(index=False))
else:
    print("No customer needs data found.\n")
````
