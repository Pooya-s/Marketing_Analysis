```python
from time import sleep
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import os

# Setup Selenium WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
service = Service(r'c:\\Users\\pooya\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe')  # Update with your ChromeDriver path

driver = webdriver.Chrome(service=service, options=chrome_options)

# Data placeholders
users = []
locations = []
dates = []
ratings = []
reviews = []
titles = []
experience_dates = []

# Configuration
company_url = 'https://it.trustpilot.com/review/sorgenia.it'
max_pages = 574  # Set the maximum number of pages to scrape
current_page = 1

temp_file = 'Sorgenia_reviews_temp.csv'

# Load progress if available
if os.path.exists(temp_file):
    temp_data = pd.read_csv(temp_file)
    users = temp_data['Username'].tolist()
    locations = temp_data['Location'].tolist()
    dates = temp_data['Date'].tolist()
    reviews = temp_data['Content'].tolist()
    ratings = temp_data['Rating'].tolist()
    titles = temp_data['Title'].tolist()
    experience_dates = temp_data['Experience_date'].tolist()
    current_page = len(temp_data) // len(users) + 1

try:
    while current_page <= max_pages:
        url = f"{company_url}?page={current_page}" if current_page > 1 else company_url
        print(f"Scraping page {current_page}: {url}")

        max_retries = 5
        retry_count = 0

        while retry_count < max_retries:
            try:
                driver.get(url)
                sleep(3)  # Allow time for the page to load
                break  # Exit the retry loop on success
            except Exception as e:
                retry_count += 1
                print(f"Error loading page {current_page}: {e}. Retrying {retry_count}/{max_retries}...")
                sleep(5)

        if retry_count == max_retries:
            print(f"Failed to load page {current_page} after {max_retries} retries. Skipping.")
            with open('scraping_errors.log', 'a') as log_file:
                log_file.write(f"Failed to scrape page {current_page}\n")
            current_page += 1
            continue

        # Scroll to the bottom to load all reviews (if lazy-loaded)
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        sleep(2)  # Allow time for additional content to load

        # Extract reviews
        review_elements = driver.find_elements(By.CSS_SELECTOR, '.styles_reviewCardInner__EwDq2')  # Update this selector
        if not review_elements:
            print(f"No more reviews found on page {current_page}. Stopping.")
            break

        for review in review_elements:
            try:
                user = review.find_element(By.CSS_SELECTOR, '.typography_heading-xxs__QKBS8.typography_appearance-default__AAY17').text  # Update this selector
                content = review.find_element(By.CSS_SELECTOR, '.styles_reviewContent__0Q2Tg > p:first-of-type').text
                title = review.find_element(By.CSS_SELECTOR, '.styles_reviewContent__0Q2Tg > a').text
                date = review.find_element(By.CSS_SELECTOR, 'time').text  # Update this selector
                rating = review.find_element(By.CSS_SELECTOR, '.star-rating_starRating__4rrcf img').get_attribute('alt')
                rating = rating.split(' ')[1]  # Extract the rating value (e.g., "5")  # Update this attribute
                experience_date_element = review.find_element(By.CSS_SELECTOR, '.styles_reviewContent__0Q2Tg > p[data-service-review-date-of-experience-typography="true"]')
                experience_date = experience_date_element.text.split(': ')[1]

                # Conditional to check if experience_date is in content
                if experience_date in content:
                    content = 'N/A'

                users.append(user)
                reviews.append(content)
                dates.append(date)
                ratings.append(rating)
                titles.append(title)
                experience_dates.append(experience_date)

                # Location is optional
                try:
                    location = review.find_element(By.CSS_SELECTOR, '.typography_body-m__xgxZ_.typography_appearance-subtle__8_H2l.styles_detailsIcon__Fo_ua > span').text  # Update this selector
                except:
                    location = 'Unknown'

                locations.append(location)

            except Exception as e:
                print(f"Error extracting review: {e}")
                continue

        # Save progress after each page
        review_data = pd.DataFrame({
            'Username': users,
            'Location': locations,
            'Date': dates,
            'Content': reviews,
            'Rating': ratings,
            'Title': titles,
            'Experience_date': experience_dates
        })
        review_data.to_csv(temp_file, index=False)

        current_page += 1
        sleep(2)  # Delay before loading the next page

finally:
    driver.quit()

# Save final output
review_data.to_csv('Sorgenia_reviews.csv', index=False)
print("Reviews saved to Sorgenia_reviews.csv")
```
